```python
# import the libraries that we will need.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#Load the dataset into a Pandas DataFrame.
import pandas as pd
dataset = pd.read_csv('Downloads/HRDataset_v14.csv')
```

```python
#provide information about data type
dataset.info()
```

```python
#display the top 10 rows from the dataset
dataset.head(10)
```

```python
#check whether any column contain null values or not
dataset.isnull().sum()
```

```python
#fill null values with 0
dataset.fillna("0",inplace=True)
#drop the duplicate values
dataset.drop_duplicates(inplace=True)
dataset.drop_duplicates(subset='EmpID', keep='first', inplace=True)
```

```python
#again check wheather any column contain any null values
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
# sort the values of employee id in the dataset
dataset.sort_values('EmpID',ascending=True,inplace=True)
dataset.Salary.sort_values(ascending=False).head(10)
```

```python
#drop the columns which are not needed
dataset.drop(['MaritalStatusID', 'MarriedID','GenderID','DaysLateLast30','RaceDesc','Zip','State','TermReason','HispanicLatino','CitizenDesc','MaritalDesc'], axis=1)
```

```python
#prepare our data by calculating the age of employees and ensuring data types are correct.
# Convert 'DOB' to datetime and calculate age
dataset['DOB'] = pd.to_datetime(dataset['DOB'])
current_year = datetime.now().year
dataset['Age'] = current_year - dataset['DOB'].dt.year

# Convert relevant columns to appropriate data types
dataset['PerfScoreID'] = dataset['PerfScoreID'].astype(int)
dataset['Salary'] = dataset['Salary'].astype(int)
dataset['PositionID'] = dataset['PositionID'].astype(int)
```

```python
# create scatter plots to visualize the relationships between performance score and salary, age, and job level.
# Performance vs. Salary
plt.figure(figsize=(10, 5))
sns.scatterplot(data=dataset, x='PerfScoreID', y='Salary')
plt.title('Performance Score vs Salary')
plt.show()
```

```python
# Performance vs. Age
plt.figure(figsize=(10, 5))
sns.scatterplot(data=dataset, x='PerfScoreID', y='Age')
plt.title('Performance Score vs Age')
plt.show()
```

```python
# Performance vs. Job Level
plt.figure(figsize=(10, 5))
sns.scatterplot(data=dataset, x='PerfScoreID', y='PositionID')
plt.title('Performance Score vs Job Level')
plt.show()
```

```python
# Calculating correlation coefficients
salary_corr = dataset['PerfScoreID'].corr(dataset['Salary'])
age_corr = dataset['PerfScoreID'].corr(dataset['Age'])
job_level_corr = dataset['PerfScoreID'].corr(dataset['PositionID'])

# Print the correlation coefficients
print(f"Correlation between Performance Score and Salary: {salary_corr}")
print(f"Correlation between Performance Score and Age: {age_corr}")
print(f"Correlation between Performance Score and Job Level: {job_level_corr}")
```

```python
# Grouping by Department and calculating average satisfaction
department_satisfaction = dataset.groupby('Department')['EmpSatisfaction'].mean()

# Creating a bar chart with figure size specified for better visibility
plt.figure(figsize=(12, 6))
bars = plt.bar(department_satisfaction.index, department_satisfaction.values)

# Adding the value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.title('Average Employee Satisfaction by Department')
plt.ylabel('Average Satisfaction')
plt.xticks(rotation=45, ha='right')  # Rotate labels to avoid overlap
plt.show()
```

```python
# Calculating attrition rate by job level
attrition_rate = dataset[dataset['Termd'] == 1].groupby('PositionID').size() / dataset.groupby('PositionID').size()

# Creating a bar chart
plt.figure(figsize=(10, 5))
attrition_rate.plot(kind='bar')
#give title to the graph
plt.title('Attrition Rate by Job Level')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Task 1: Measure of Central Tendency

# Calculate Mean, Median, and Mode of Salary for PerfScoreID
PerfScoreID = dataset['PerfScoreID'].unique()
for  level in PerfScoreID:
    Salary_for_PerfScoreID = dataset[dataset['PerfScoreID'] == level]['Salary']
    mean_balance = np.mean(Salary_for_PerfScoreID)
    median_balance = np.median(Salary_for_PerfScoreID)
    mode_balance = Salary_for_PerfScoreID.mode().values[0]  # Mode might have multiple values, picking the first

    print(f"PerfScoreID: {level}")
    print(f"Mean Salary: {mean_balance:.2f}")
    print(f"Median Salary: {median_balance:.2f}")
    print(f"Mode Salary: {mode_balance}\n")
```

```python
# Task 2: Measure of Spread/Dispersion for Salary for IT/IS Department
itis = dataset[dataset['Department'] == 'IT/IS']['Salary']
range_balance = np.max(itis) - np.min(itis)
variance_balance = np.var(itis)
std_dev_balance = np.std(itis)

print(f"Range of Salary for IT/IS job: {range_balance:.2f}")
print(f"Variance of Salary for IT/IS job: {variance_balance:.2f}")
print(f"Standard Deviation of Salary  for IT/IS job: {std_dev_balance:.2f}")
```

```python
#task 3 : Calculate pmf dor Dept ID
dept_id_counts = dataset['DeptID'].value_counts()

# Total number of observations
total_observations = dept_id_counts.sum()

# Calculate probabilities for each 'DeptId'
pmf = dept_id_counts / total_observations
print(pmf)
# Plot PMF
plt.bar(pmf.index, pmf.values, align='center', alpha=0.7)
plt.xlabel('DeptId')
plt.ylabel('Probability')
plt.title('PMF for DeptId')
plt.show()
```

```python
#Task 4 : PDF
from scipy.stats import gaussian_kde

# Estimating PDF for 'Salary'
kde = gaussian_kde(dataset['Salary'])
income_values = np.linspace(dataset['Salary'].min(), dataset['Salary'].max(), 100)
pdf_income = kde(income_values)

plt.plot(income_values, pdf_income)
plt.xlabel('Salary')
plt.ylabel('Density')
plt.title('Estimated PDF for Salary')
plt.show()
```
